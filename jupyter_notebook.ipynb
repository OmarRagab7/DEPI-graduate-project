# Cleaning  Dataset 

```python
df.dropna(how='all', inplace=True)
```

```python
df.info()
```

## Check for Missing Values using Interquartile Range (IQR)

```python
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nOutliers detected using IQR:")
print(outliers)
```

## Boxplot for Columns with Outliers

```python
outlier_columns = outliers[outliers > 0].index
for col in outlier_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}", fontsize=14, fontweight='bold')
    plt.show()
```

# Data Analysis using Group By and Pivot Table

```python
print("\nEmployee Count by Department:")
print(df.groupby('DepartmentType')['EmpID'].count())
```

```python
print("\nAverage Salary by Department:")
print(df.pivot_table(index='DepartmentType', values='Desired_Salary', aggfunc='mean'))
```

# Important Charts

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='DepartmentType', data=df, hue=None, palette='viridis')
plt.title('Employee Count per Department', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='DepartmentType', y='Desired_Salary', data=df, palette='coolwarm')
plt.title('Salary Distribution by Department', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```
